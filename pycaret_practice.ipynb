```python
import os
import pandas as pd
import numpy as np

PATH = "../input/kagglecamp2022"
train = pd.read_csv(os.path.join(PATH, "train.csv"))
```

```python
train.head()
```

```python
from pycaret import classification as pyc_clf
```

```python
clf = pyc_clf.setup(train, 
                    train_size=0.8,
                    fold = 5,
                    target="Cover_Type",
                    session_id = 42, # 랜덤스테이트랑 같은 값
                    categorical_features=["Soil_Type", "Wilderness_Area"]
                   )
```

```python
pyc_clf.models()
```

```python
dt = pyc_clf.create_model('dt')
```

```python
dt
```

```python
top_3 = pyc_clf.compare_models(include=['dt','rf','lightgbm'], n_select=3)
```

```python
tune_dt = pyc_clf.tune_model(top_3[2], n_iter=4)
```
